```python
locations_remove = ['LITTLE BOW LAKE RESERVOIR', 'CHAIN LAKES RESERVOIR', 'SPRAY LAKES RESERVOIR', 'THINLAKE RIVER', 
                    'MOOSELAKE RIVER', 'PIGEON LAKE CREEK', 'BUCK LAKE CREEK']

subset_df = df.loc[df['waterbody_name'].str.lower().str.contains('lake')]
subset_df = subset_df.loc[~subset_df['waterbody_name'].isin(locations_remove)]
print(len(subset_df))
display(subset_df.head())
```

```python
lakes.pivot(index='waterbody_name', columns='decade', values='decade')
```

```python
df_decades = pd.get_dummies(lakes, columns=['decade'], prefix='', prefix_sep='')
df_decades.replace(0, np.nan, inplace=True)
for col in df_decades.columns[4:]:
    df_decades[col] = df_decades[col].fillna(df_decades.groupby('waterbody_name')[col].transform('mean'))

df_decades.drop_duplicates(inplace=True)
df_decades['how_many_decades_sampled'] = df_decades[df_decades.columns[4:]].sum(axis=1)
df_decades
```
