```python
alberta_dfs = []

path = '/mnt/c/Users/imrit/Downloads/Sharma_fish_project/GET_WATERBODY_NAMES/ALBERTA_TO_GET_WATERBODIES/'

for file in os.listdir(path):
    df_temp = pd.read_csv(f'{path}{file}')
    alberta_dfs.append(df_temp)
    
alberta = pd.concat(alberta_dfs)
alberta = alberta[['latitude', 'longitude', 'waterbody_name', 'site_description']]
alberta['waterbody_type'] = alberta['waterbody_name'].str.extract("(" + "|".join(waterbody_type_list) +")", expand=False)
alberta
```

```python
df_alberta_unidentified = df.copy(deep=True).merge(alberta, on = ['latitude', 'longitude'], how = 'left')
len(df_alberta_unidentified)
```
